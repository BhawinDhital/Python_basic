### Seaborn: statistical data visualization
Seaborn is a Python data visualization library based on matplotlib. It provides a
high-level interface for drawing attractive and informative statistical graphics. It also
is designed to work very well with pandas dataframe objects.

# Distribution plots

```python
import seaborn as sns
import numpy as np
```

```python
%matplotlib inline
```

```python
sns.set(); np.random.seed(0)
x=np.random.randn(100)
ax = sns.distplot(x)
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.distplot(tips['total_bill'])
# distplot function combines the matplotlib hist function ( with automatic calculation of
# a good default bin size) with the seaborn kdeplot and
#rugplot() functions.
```

```python
sns.distplot(tips['total_bill'],kde=False)
# This removes the line as shown in the previous figure! Just a histogram
```

```python
# You can change the bins value
sns.distplot(tips['total_bill'],kde=False,bins=30)
# Most of your total_bills is mostly betwwen 10 dollar to 20 dollar
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=100)
```

```python
# Jointplot for bi-variate data means two variables
sns.jointplot(x='total_bill',y='tip',data=tips)
# x axis is total_bill and y axis is tip.Load from data tips.
```

```python
# you can plot hexagonal distribution plot using hex
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
# more points, it gets darker in the histogram, less point it gets lighter. 
```

```python
# Regression plot
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
# Pairplot plots pairwise relationship across the dataframe
sns.pairplot(tips)
```

```python
# hue arguments for categorical
sns.pairplot(tips,hue='sex')
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```python
# seaborn.rugplot plots the datapoints in an aray as sticks on an axis.
sns.rugplot(tips['total_bill'])
```

```python
# kde means kernel density estimation
sns.kdeplot(tips['total_bill'])
```

# Categorical Plots

# barplot
It shows point estimates and confidence intervals as rectangular bars. A bar plot represnts an
estimate of central tendency for a numeric variable with the height of each rectangle and provides some indication 
of the uncertainty aroud that estimate using error bars.

```python
sns.barplot(x='sex',y='total_bill',data=tips)
# shows the average of the total_bill
```

```python
# estimator is a function, statistical function to estimate within each categorical bins.
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
# This gives the standard deviation
```

```python
# countplot is same as barplot, shows the counts of observations in each categorical bins using bars.
sns.countplot(x='sex',data=tips)
```

```python
# boxplot
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
# violinplot looks like violin
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
#stripplot which has one variable as a categorical and another numeric as before
sns.stripplot(x='day',y='total_bill',data=tips)
# here it is overlapping and difficult to distinguish the data, top of each otherr
# to give a little noise add jitter
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',split=True)
```

```python
# swarmplot() which is beeswarm. This is a combination of stripplot and violinplot
# Not good for very large dataset
sns.swarmplot(x = 'day',y='total_bill',data=tips)
```

```python
# Combiing both violinplot and swarmplot
sns.violinplot(x = 'day',y='total_bill',data=tips)
sns.swarmplot(x = 'day',y='total_bill',data=tips,color='black')
```

```python
# factorplot is a general for all of these. Now factorplot is  modified to catplot
sns.catplot(x = 'day',y ='total_bill',data=tips,kind='bar')
```

```python
sns.catplot(x = 'day',y ='total_bill',data=tips,kind='violin')
```

# Matrix Plots

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
tips.head()
```

```python
flights.head()
```

```python
# Correlation data
tc = tips.corr()
tc
```

heatmap is a two-dimensional grphical representation of data where the indiviual values that are contained
in a matrix are representated as colors.The seaborn python package allows the creation of annotated heatmaps
which can be pulled using matplotlib tools as per the creator's requirement.

```python
#sns.heatmap plot rectangular data as a color-encoded matrix.
sns.heatmap(tc)
```

```python
sns.heatmap(tc,annot=True)
```

```python
# cmap=colormap
sns.heatmap(tc,annot=True,cmap='coolwarm')
```

```python
flights
```

```python
fp = flights.pivot_table(index='month',columns='year',values='passengers')
# This command gives the output in the matriix form
#columns=year means row in the output
fp
```

```python
sns.heatmap(fp)
```

```python
# You can change the color using cmap
sns.heatmap(fp,cmap='magma')
```

```python
# line color and line width
sns.heatmap(fp,cmap='magma',linecolor='white',linewidths=1)
```

```python
sns.heatmap(fp,cmap='coolwarm',linecolor='white',linewidths=1)
```

```python
sns.heatmap(fp)
```

```python
#clustermap
sns.clustermap(fp)
```

```python
sns.clustermap(fp,cmap='coolwarm')
```

# Grids

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
g = sns.FacetGrid(data = tips,col='time',row='smoker')
g.map(sns.distplot,'total_bill')
```

```python
g = sns.FacetGrid(data = tips,col='time',row='smoker')
g.map(plt.scatter,'total_bill','tip')
```

# Regression Plots
The regression plot in seaborn are primarily intended toadd a visual guide that helps to emphasize paterns in a dataset during exploratory data analysis. It in general defines a regression line between 2 parameters and helps to visualize their linear relationships.

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'])
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],
          scatter_kws={'s':100})
# plotting the bigger size
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex')
#separate plot for sex showing Male and Female.
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',
          aspect=0.6,height=8)
#height was before size. aspect ratio to control the width of the plot.
```

# Style and color

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.set_style('white')# I creates white backgroung
sns.countplot(x = 'sex',data=tips)
```

```python
sns.set_style('ticks')# I creates ticks at the edge
sns.countplot(x = 'sex',data=tips)
```

```python
sns.set_style('whitegrid')
sns.countplot(x = 'sex',data=tips)
```

```python
sns.set_style('ticks')
sns.countplot(x = 'sex',data=tips)
sns.despine(left=True,bottom=True)
```

```python
# You can use matplotlib along with seaborn to control the figure size
plt.figure(figsize=(12,3))
sns.countplot(x = 'sex',data=tips)
```

```python
sns.set_context('poster') # poster gives large size
sns.countplot(x = 'sex',data=tips)
```

```python
# If you want to increase the size then use font_scale=
sns.set_context('poster',font_scale=3)
sns.countplot(x = 'sex',data=tips)
```

```python
sns.set_context('notebook')
sns.countplot(x = 'sex',data=tips)
```
