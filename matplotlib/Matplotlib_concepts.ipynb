# Matplotlib Overview 

Matplotlib is the "grandfather" library of data visualization with Python. It was created by John Hunter. 
Matplotlib allows you to create reproducible figures programmatically.

```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
y
```

```python
# Functional method / use shift+tab to see several arguments
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'r-')# red color
```

```python
plt.plot(x,y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
```

```python
plt.subplot(1,2,1) # for two subplots in the same place
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object oriented method
fig = plt.figure()
```

```python
fig = plt.figure()# This method first creat the figure background
axes = fig.add_axes([0.1,0.1,0.8,0.8])# add axes on left,bottom, width and height
axes.plot(x,y)# same plot as earlier only this is object oriented method.
axes.set_xlabel('X label')
axes.set_ylabel('Y Label')
axes.set_title(' Set title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])# 10% from the left of the figure,10% from the bottom of the figure,
#80% width of the figure,80% height of the figure
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
#This second axes is created and placed 20% from the left of the figure, 50% from the bottom of the figure,
#takes up 40% of the width of the figure, and takes up 30% of the height of the figure.
#This places this axes within the larger axes
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes1.set_title('Larger plot')
axes2.plot(y,x)
axes2.set_title('Smaller plot')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title('Title')
ax.set_xlabel('X')
ax.set_ylabel('Y')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.plot(y,x)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='X Squared')
ax.plot(x,x**3, label='X Cubed')
ax.legend()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='X Squared')
ax.plot(x,x**3, label='X Cubed')
ax.legend(loc=(0.1,0.1))# go to pyplot and see loc
```

# Plot Appearance 

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='orange')#RGB Hex code search online for different color
```

```python
# Line width and line style
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='orange',linewidth=20)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='orange',linewidth=3,alpha=0.5)#alpha gives the transparency,linewidth=lw also works
```

```python
# Line style
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='orange',lw=3,linestyle=':')
```

```python
# Line style
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='orange',lw=3,linestyle='steps')
```

```python
# Line style
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='orange',lw=3,linestyle='--')# linestyle=ls also works
```

```python
# Markers
x
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='orange',lw=3,ls='-',marker='o')# change marker='*'etc
```

```python
# Marker size
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='orange',lw=3,ls='-',marker='+',markersize=20)
```

```python
# Marker face color
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='orange',lw=3,ls='-',marker='o',markersize=20,
       markerfacecolor='yellow',markeredgecolor='green',markeredgewidth=3)
# Go to line and marker style in pyplot in matplotlib webpage
```

```python
#control over axis apperance and plot range
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',lw=2,ls='--')
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
