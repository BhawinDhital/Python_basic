# Histogram

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
mal_age = np.random.randint(18,45,(100))
fem_age = np.random.randint(15,40,(100))
```

```python
print(mal_age)
print(fem_age)
```

```python
# to generate histogram
```

```python
plt.hist(mal_age)
plt.show()
```

```python
plt.hist(mal_age)
plt.title('Male student age')
plt.xlabel('age of teh student')
plt.ylabel('frequency')
plt.show()
```

```python
# to provide the bin number such that the number between 2 & 15 will be in 1 bar
# 20 and 25 in another bar
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(mal_age,bins)
plt.title('male student age')
plt.xlabel('age of the student')
plt.ylabel('frequency')
plt.show()
```

```python
# to show the bar separation
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(mal_age,bins,rwidth = 0.8) # shows 80 %space and 20 % space empty
plt.title('male student age')
plt.xlabel('age of the student')
plt.ylabel('frequency')
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(mal_age,bins,rwidth = 0.8,histtype = 'step') # shows 80 %space and 20 % space empty
plt.title('male student age')
plt.xlabel('age of the student')
plt.ylabel('frequency')
plt.show()
```

```python
# histogram type
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(mal_age,bins,rwidth = 0.6,histtype = 'bar') # shows 80 %space and 20 % space empty
plt.title('male student age')
plt.xlabel('age of the student')
plt.ylabel('frequency')
plt.show()
```

```python
# to change the orientations of the histogram
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(mal_age,bins,rwidth = 0.6,histtype = 'bar',orientation = 'horizontal') # orientation is horizontal
plt.title('male student age')
plt.xlabel('age of the student')
plt.ylabel('frequency')
plt.show()
```

```python
# to change the color of the bar
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(mal_age,bins,rwidth = 0.6,histtype = 'bar',orientation = 'horizontal',color = 'b') # orientation is horizontal
plt.title('male student age')
plt.xlabel('age of the student')
plt.ylabel('frequency')
plt.show()
```

```python
# lets give the legend
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(mal_age,bins,rwidth = 0.6,histtype = 'bar',orientation = 'horizontal',color = 'b', label = ' ML student') # orientation is horizontal
plt.title('male student age')
plt.xlabel('age of the student')
plt.ylabel('frequency')
plt.legend(loc=4)
plt.show()
```

```python
# two data set plot in the same plot
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(mal_age,bins,rwidth = 0.6,histtype = 'bar',orientation = 'horizontal',color = 'b', label = ' ML student') # orientation is horizontal
plt.hist(fem_age,bins,rwidth = 0.6,histtype = 'bar',orientation = 'horizontal',color = 'r', label = ' FL student')
plt.title('male student age')
plt.xlabel('age of the student')
plt.ylabel('frequency')
plt.legend(loc=4)
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(mal_age,bins,rwidth = 0.6,histtype = 'bar',orientation = 'vertical',color = 'b', label = ' ML student')
plt.title('male student age')
plt.xlabel('age of the student')
plt.ylabel('frequency')
plt.legend() # no more location
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist([mal_age,fem_age],bins, rwidth = 0.6,histtype = 'bar', orientation = 'vertical',
         color = ['b','g'], label = ['ML student','FL student']) # orientation is horizontal
plt.title('male student age')
plt.xlabel('age of the student')
plt.legend()
plt.ylabel('frequency')
plt.show()
```

```python
# to change the size of the histogram
```

```python
bins = [15,20,25,30,35,40,45]
plt.figure(figsize=(16,9))
plt.hist([mal_age,fem_age],bins, rwidth = 0.6,histtype = 'bar', orientation = 'vertical',
         color = ['b','g'], label = ['ML student','FL student']) # orientation is horizontal
plt.title('male student age')
plt.xlabel('age of the student')
plt.legend()
plt.ylabel('frequency')
plt.show()
```

# Bar chart

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
classes = ['python','R','AI','ML','DS']
class1 = [30,10,20,25,10]
class2 = [40,5,20,20,10]
class3 = [35,5,30,15,15]
```

```python
plt.bar(classes,class1)
```

```python
# to plot horizontal use barh
```

```python
plt.barh(classes,class1) #bar horizontal(barh)
```

```python
plt.bar(classes,class1,width = 0.2)
```

```python
plt.bar(classes,class1,width = 0.2, align = 'edge')
```

```python
plt.bar(classes,class1,width = 0.2, align = 'edge',color = 'b') # color
```

```python
plt.bar(classes,class1,width = 0.2, align = 'edge',color = 'g',edgecolor = 'm',linewidth = 5)
```

```python
# change the opacity or make it transparent
```

```python
plt.bar(classes,class1,width = 0.2, align = 'edge',color = 'g',edgecolor = 'm',linewidth = 5,alpha=0.5)
```

```python
# line style
plt.bar(classes,class1,width = 0.2, align = 'edge',color = 'g',edgecolor = 'm',linewidth = 5,alpha=0.5,
       linestyle = '--',label = 'class1 students')
```

```python
# hide the histogram
# line style
plt.bar(classes,class1,width = 0.2, align = 'edge',color = 'g',edgecolor = 'm',linewidth = 5,alpha=0.5,
       linestyle = '--',label = 'class1 students',visible = False)
```

```python
style.use("ggplot")
# line style
plt.bar(classes,class1,width = 0.2, align = 'edge',color = 'g',edgecolor = 'm',linewidth = 5,alpha=0.5,
       linestyle = '--',label = 'class1 students')
```

```python
plt.figure(figsize=(15,9))
# line style
plt.bar(classes,class1,width = 0.2, align = 'edge',color = 'g',edgecolor = 'm',linewidth = 5,alpha=0.5,
       linestyle = '--',label = 'class1 students')
```

```python
plt.figure(figsize =(15,9))
plt.bar(classes,class1, width = 0.6, align ="edge", color = 'k',
       edgecolor='m', linewidth=5,alpha=0.6, linestyle = "--", label = "class1  students")

plt.bar(classes,class2, width = 0.6, align ="edge", color = 'r',
       edgecolor='k', linewidth=5,alpha=0.6, linestyle = "--", label = "class2 students")


plt.bar(classes,class3, width = 0.6, align ="edge", color = 'g',
       edgecolor='y', linewidth=5,alpha=0.6, linestyle = "--", label = "class3 students")

plt.title("Bar chart of AIP class",fontsize =50)
plt.xlabel("classes", fontsize=50)
plt.ylabel("frequency", fontsize=50)
plt.show()
```

```python
plt.figure(figsize =(15,9))
classes_index=np.arange(len(classes)) # new line
width = 0.2
plt.bar(classes_index,class1, width, align ="edge", color = 'k',
        label = "class1 students")

plt.bar(classes_index+2*width,class2, width, align ="edge", color = 'r',
        label = "class2 students")


plt.bar(classes_index+width,class3, width, align ="edge", color = 'g',
        label = "class3 students")

plt.title("Bar chart of AIP class",fontsize =50)
plt.xlabel("classes", fontsize=50)
plt.ylabel("frequency", fontsize=50)
plt.show()
```
